本レポートでは手書きの数字の画像データを使い、機械学習のモデルの学習と予測を行います。
SVMとロジスティック回帰について機械学習を行います。

具体的には下図のようなデータを用います。

```python
from sklearn import datasets
digits = datasets.load_digits()

# データの確認
import matplotlib.pyplot as plt
plt.matshow(digits.images[0], cmap="Greys")
plt.show()
```

```python
X = digits.data
y = digits.target

# 訓練データ ：偶数行
X_train, y_train = X[0::2], y[0::2]
# テストデータ：奇数行
X_test, y_test = X[1::2], y[1::2]
```

**SVMについて機械学習**

```python
# SVM
from sklearn import svm
clf = svm.SVC(gamma=0.001)
clf.fit(X_train, y_train)
```

```python
from sklearn import metrics

# テストデータで試した正解率を返す
accuracy = clf.score(X_test, y_test)
print(f"正解率{accuracy}")

# 学習済モデルを使ってテストデータを分類した結果を返す
predicted = clf.predict(X_test)

# precision(適合率): 
# recall(再現率) : 
# F-score(F値) : 
print("classification")
print()
print(metrics.classification_report(y_test, predicted))
```

**ロジスティック回帰について機械学習**

```python
# ロジスティック回帰
from sklearn.linear_model import LogisticRegression
clf = LogisticRegression()
clf.fit(X_train, y_train)
```

```python
from sklearn import metrics

# テストデータで試した正解率を返す
accuracy = clf.score(X_test, y_test)
print(f"正解率{accuracy}")

# 学習済モデルを使ってテストデータを分類した結果を返す
predicted = clf.predict(X_test)

# precision(適合率): 
# recall(再現率) : 
# F-score(F値) : 
print("classification")
print()
print(metrics.classification_report(y_test, predicted))
```

### **考察**

それぞれの学習の結果、SVMを適用した方が高い正答率であることがわかった。
また適合率、再現率、F値の三点からも学習の比較を試みた結果、いずれもSVMの方が高い数値を出している。よって今回の場合、SVMの方がより適した学習モデルであると言える。
